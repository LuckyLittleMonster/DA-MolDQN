```python
def mol_with_atom_index_test( mol ):
    atoms = mol.GetNumAtoms()
    for idx in range( atoms ):
        # pass
        if mol.GetAtomWithIdx( idx ).GetIdx() == 16 or mol.GetAtomWithIdx( idx ).GetIdx() == 19:
            mol.GetAtomWithIdx( idx ).SetProp( 'molAtomMapNumber', str( mol.GetAtomWithIdx( idx ).GetIdx() ) )
    return mol
def mol_with_atom_index( mol, ids = None ):
    atoms = mol.GetNumAtoms()
    for idx in range( atoms ):
        # pass
        # if mol.GetAtomWithIdx( idx ).GetIdx() == 16 or mol.GetAtomWithIdx( idx ).GetIdx() == 19:
        if ids is None or (idx in ids):
            mol.GetAtomWithIdx( idx ).SetProp( 'molAtomMapNumber', str( idx ) )
    return mol
```

```python
df_good = pd.read_csv('/data/gpfs/Users/l1062811/git/RL4-working/good_smiles_ip_env_tem.txt', sep='\t')
df_good = df_good.sort_values(df_good.columns[4])
df_err = pd.read_csv('/data/gpfs/Users/l1062811/git/RL4-working/error_smiles_ip_env_tem.txt', sep='\t')
df_err = df_err.sort_values(df_err.columns[4])

lg = len(df_good.iloc[:, 0].values)
g = []
s = 0.0
for i, b, ab in zip(range(lg), df_good.iloc[:, 4].values, df_good.iloc[:, 5].values):
    g.append(i * 1.0 / lg);
    s += ab

le = len(df_err.iloc[:, 0].values)
e = []
for i, b, ab in zip(range(le), df_err.iloc[:, 4].values, df_err.iloc[:, 5].values):
    e.append(i * 1.0 / le);
    s += ab
s = s / (lg + le )

plt.figure(figsize=(5,5))
init_l = plt.axhline(s, color='orange')

pred_g = plt.scatter(g, df_good.iloc[:, 4].values, color='red', s = 2)
pred_e = plt.scatter(e, df_err.iloc[:, 4].values, color='blue', s = 2)
# plt.xlim((100, 190))
# plt.ylim((100, 190))
plt.xlabel('Percentage')
plt.ylabel('Distance')
plt.title('The distance between atoms that the new bond is added to')
plt.legend(handles=[pred_g, pred_e, init_l],labels=['Valid', 'Invalid', 'Average distance'],loc='best')

plt.savefig('distance_all.png', dpi = 500)
```

```python
df_good = pd.read_csv('/data/gpfs/Users/l1062811/git/RL4-working/good_smiles_ip_env_tem.txt', sep='\t')
df_good = df_good.sort_values(df_good.columns[4])
df_err = pd.read_csv('/data/gpfs/Users/l1062811/git/RL4-working/error_smiles_ip_env_tem.txt', sep='\t')
df_err = df_err.sort_values(df_err.columns[4])
df_good.columns =['init_smiles', 'gm', 'bid', 'eid', 'bond_dis', 'ave_dis']
df_err.columns =['init_smiles', 'gm', 'bid', 'eid', 'bond_dis', 'ave_dis']
df_good.set_index('init_smiles', inplace=True)
df_err.set_index('init_smiles', inplace = True)

mols = ['CN(C)c1ccc(O)c(N(C)C)n1',
       'Cc1c(O)cnc(N(C)CN)c1N(C)C(N)O',
       'CC(=N)N(CO)c1c(O)cnc(C)c1N=O',
       'CN(C)c1cc(O)cnc1N(C)C']
mol_id = 2
mol = mols[mol_id]

dfg = df_good.loc[mol]
dfe = df_err.loc[mol]
# dfg.sort_values(by=['bond_dis'])
# dfe.sort_values(by=['bond_dis'])
# print(dfe.init_smiles.values)
lg = len(dfg.iloc[:, 0].values)
# print(lg)
g = []
s = 0.0
for i, b, ab in zip(range(lg), dfg.bond_dis.values, dfg.ave_dis.values):
    g.append(i * 1.0 / lg);
    s += ab

le = len(dfe.iloc[:, 0].values)
e = []
for i, b, ab in zip(range(le), dfe.bond_dis.values, dfe.ave_dis.values):
    e.append(i * 1.0 / le);
    s += ab
s = s / (lg + le )

plt.figure(figsize=(5,5))
init_l = plt.axhline(s, color='orange')

pred_g = plt.scatter(g, dfg.bond_dis.values, color='red', s = 5)
pred_e = plt.scatter(e, dfe.bond_dis.values, color='blue', s = 5)
plt.xlabel('Percentage')
plt.ylabel('Distance')
plt.title('The distance between atoms (Mol: {})'.format(mol_id))
plt.legend(handles=[pred_g, pred_e, init_l],labels=['Valid', 'Invalid', 'Average distance'],loc='best')

plt.savefig('distance_{}.png'.format(mol_id), dpi = 500)
```
