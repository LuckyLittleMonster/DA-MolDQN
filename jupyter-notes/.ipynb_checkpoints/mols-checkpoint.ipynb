```python
import cairosvg
import io
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import pickle
import random
import rdkit
import sys
from rdkit import Chem
from rdkit import RDConfig
from rdkit.Chem import Draw
from rdkit.Chem import QED
from rdkit.Chem import rdFMCS
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem.Draw import rdDepictor
from rdkit.Chem.Draw import rdMolDraw2D
from rdkit.DataManip.Metric import GetTanimotoDistMat
from rdkit.DataManip.Metric import GetTanimotoSimMat
from IPython.display import Image, display
import tools
from tools import Results
from glob import glob

%load_ext autoreload
%autoreload 2

IPythonConsole.ipython_useSVG=True
sys.path.append(os.path.join(RDConfig.RDContribDir, 'SA_Score'))
import sascorer

mol_path_dir = '../results/path'

def mol_to_png(mol, file_name, dir_name = '../results/path', highlightAtoms = None, highlightAtomColors = None, l = '', width=300, height=200):
    full_path = f'{dir_name}/{file_name}.png'
    drawer = rdMolDraw2D.MolDraw2DSVG(width, height)
    drawer.DrawMolecule(mol, highlightAtoms=highlightAtoms, highlightAtomColors=highlightAtomColors, legend=l)
    drawer.FinishDrawing()
    cairosvg.svg2png(bytestring=drawer.GetDrawingText().encode(), write_to=full_path)
    return full_path

def draw_with_diff(mol1, mol2, file_name, l = ''):
    mcs = rdFMCS.FindMCS([mol1,mol2])
    mcs_mol = Chem.MolFromSmarts(mcs.smartsString)
    
    if mol1.GetNumAtoms() <= mol2.GetNumAtoms():
        # highlight the newly added bond or atom, 
        match2 = mol2.GetSubstructMatch(mcs_mol)
        target_atm2 = []
        for atom in mol2.GetAtoms():
            if atom.GetIdx() not in match2:
                target_atm2.append(atom.GetIdx())
        return mol_to_png(mol2, file_name, highlightAtoms = target_atm2, l = l)
    else:
        # removed a bond or an atom
        match1 = mol1.GetSubstructMatch(mcs_mol)
        r = Chem.RWMol(mol1)
        r.BeginBatchEdit()
        for atom in mol1.GetAtoms():
            if atom.GetIdx() in match1:
                r.RemoveAtom(atom.GetIdx())
        r.CommitBatchEdit()
        rs = Chem.MolToSmiles(r)
        m2s = Chem.MolToSmiles(mol2)
        ns = f'{m2s}.{rs}'
        nm = Chem.MolFromSmiles(ns)
        
        mcs2 = rdFMCS.FindMCS([nm,mol2])
        mcs_mol2 = Chem.MolFromSmarts(mcs2.smartsString)
        match3 = nm.GetSubstructMatch(mcs_mol2)
        target_atmn = []
        for atom in nm.GetAtoms():
            if atom.GetIdx() not in match3:
                target_atmn.append(atom.GetIdx())
        return mol_to_png(nm, file_name, highlightAtoms = target_atmn, l = l)

def draw_mols_path(mols, rewards, reward_name = 'Reward', prefix = 'test'): 
    ls = []
    for i, (m, r) in enumerate(zip(mols, rewards)):
        ls.append(f"Step: {i} {reward_name}: {r:.2f}")
    mol_to_png(mols[0], f'{prefix}_0', highlightAtoms = None, l = ls[0])
    for i in range(len(mols) - 1):
        draw_with_diff(mols[i], mols[i + 1], f'{prefix}_{i+1}', l = ls[i + 1])
    return tools.sort(glob(f'{mol_path_dir}/{prefix}_*.png'))

def draw_top_mols(p, prefix = "test"):
    ps, rs, ls = p['path'], p['reward'], [] # predicted_mols, rewards, labels
    return draw_mols_path(ps, rs, prefix = prefix)

    
print(f'rkdit version: {rdkit.__version__}')

Chem.MolFromSmiles('Cc1ccccc1O')
```

```python
# adding a 'C' to mol1
mol1 = Chem.MolFromSmiles('Cc1ccccc1O')
mol2 = Chem.MolFromSmiles('CCc1ccccc1O')
Image(filename=draw_with_diff(mol1, mol2, 'test'))
```

```python
def get_cs(mol, nm = [], aid = 0):
    c_add = (0.8, 0, 0)
    c_mod = (0, 0.8, 0.8)
    atom_cols = {}
    for i in nm:
        if i == aid:
            atom_cols[i] = c_add
        else:
            atom_cols[i] = c_mod
    return atom_cols

def mol_to_png2(mol, file_name, h = [], aid = 0, l = '', width=1200, height=600):
    full_path = f'{mol_path_dir}/{file_name}.png'
    drawer = rdMolDraw2D.MolDraw2DSVG(width, height)
    drawer.DrawMolecule(mol, highlightAtoms=h, highlightAtomColors=get_cs(mol, h, aid),legend=l)
    drawer.FinishDrawing()
    cairosvg.svg2png(bytestring=drawer.GetDrawingText().encode(), write_to=full_path)

mol1 = Chem.MolFromSmiles('c0ccccc0CC') 
mol1_path = mol_to_png(mol1, 'example_imf_1', dir_name = '../results', highlightAtoms = [0,1,3,4,5,6,7], highlightAtomColors = get_cs(mol1, [0,1,3,4,5,6,7], 7), width=1200, height=600)
display(Image(mol1_path))
mol2 = Chem.MolFromSmiles('CN(C)CCCN(C(=O)C1COc2ccccc2O1)c1nc2cc3c(cc2s1)OC(=N)O3')
mol2_path = mol_to_png(mol2, 'example_imf_2', dir_name = '../results', highlightAtoms = [30,29,28,31,24,23,25,26,22,21], highlightAtomColors = get_cs(mol2, [30,29,28,31,24,23,25,26,22,21], 30), width=1200, height=600)
display(Image(mol2_path))
```
