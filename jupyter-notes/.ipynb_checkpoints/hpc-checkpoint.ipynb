```python
path = './Experiments/hpc_11_0_episodes.pickle'
moving_average_window_size = 20
def read_data(path):
    with open(path, 'rb') as f:
        p = pickle.load(f)
        episode_time = movingaverage(p['episode_time'], moving_average_window_size)
        rewards = p['rewards']['reward']
        rewards = [np.average(r) for r in rewards]
        rewards = movingaverage(rewards, 20)
        memory = p['memory']
    return {'episode_time': episode_time, 'rewards': rewards, 'memory': memory}
# 8 bde cache
# 9 ip cache
# 10 bde + ip cache
# 11 no cache
bde = read_data('./Experiments/hpc_8_0_episodes.pickle')
ip = read_data('./Experiments/hpc_9_0_episodes.pickle')
bde_ip = read_data('./Experiments/hpc_10_0_episodes.pickle')
no_cache = read_data('./Experiments/hpc_11_0_episodes.pickle')

fig, ax = plt.subplots()
pf, = ax.plot(bde['episode_time'], label="BDE Cache")
pf, = ax.plot(ip['episode_time'], label="IP Cache")
pf, = ax.plot(bde_ip['episode_time'], label="BDE + IP Cache")
pf, = ax.plot(no_cache['episode_time'], label="No Cache")

ax.set_xlabel("Episode")
ax.set_ylabel("Computation Time (Second)")
plt.title("Computation Time (Second): Max ETKDG Attempts = 7")
plt.legend()
plt.savefig("./Results/etkdg_7_episode_time.png", dpi = 500, bbox_inches = 'tight')

plt.show()
```

```python
moving_average_window_size = 20
average_episodes = 1

plog = False
def remove_invaild_reward(r):
    if plog:
        return r
#     return r
    if r < 0.0:
        return 0.0
    if r > 3.0:
        return 0.0
    return r
def read_data(exp, trial):
    path = f'./Experiments/{exp}_{trial}_0_episodes.pickle'
    with open(path, 'rb') as f:
        p = pickle.load(f)
        episode_time = movingaverage(p['episode_time'], moving_average_window_size)
    rewards = []
    for path in glob.glob(f'./Experiments/{exp}_{trial}_*_episodes.pickle'):
        # print(path)
        with open(path, 'rb') as f:
            p = pickle.load(f)
            rank_rewards = p['rewards']['reward']
            final_reward = np.array(rank_rewards[-average_episodes:])
            final_reward = np.average(final_reward, axis=0)
            for r in final_reward:
                rewards.append(remove_invaild_reward(r))
    #     memory = p['memory']
    return {'episode_time': episode_time, 'rewards': rewards}#, 'memory': memory}
# 8 bde cache
# 9 ip cache
# 10 bde + ip cache
# 11 no cache
bde = read_data('hpc', 8)
ip = read_data('hpc', 9)
bde_ip = read_data('hpc', 10)
no_cache = read_data('hpc', 11)

fig, ax = plt.subplots()
ax.boxplot([bde['rewards'],ip['rewards'],bde_ip['rewards'],no_cache['rewards']], labels = ['BDE Cache', 'IP Cache', 'BDE + IP Cache', 'No Cache'])

ax.set_xlabel("Episode")
ax.set_ylabel("Rewards")
plt.title("Optimized Rewards: Max ETKDG Attempts = 7")
plt.savefig("./Results/etkdg_7_rewards.png", dpi = 500)
plt.show()
```

```python
path = './Experiments/hpc_11_0_episodes.pickle'
moving_average_window_size = 20
def read_data(path):
    with open(path, 'rb') as f:
        p = pickle.load(f)
        episode_time = movingaverage(p['episode_time'], moving_average_window_size)
        rewards = p['rewards']['reward']
        rewards = [np.average(r) for r in rewards]
        rewards = movingaverage(rewards, 20)
        memory = p['memory']
    return {'episode_time': episode_time, 'rewards': rewards, 'memory': memory}
# 12 bde cache
# 13 ip cache
# 14 bde + ip cache
# 15 no cache
bde = read_data('./Experiments/hpc_12_0_episodes.pickle')
ip = read_data('./Experiments/hpc_13_0_episodes.pickle')
bde_ip = read_data('./Experiments/hpc_14_0_episodes.pickle')
no_cache = read_data('./Experiments/hpc_15_0_episodes.pickle')

fig, ax = plt.subplots()
pf, = ax.plot(bde['episode_time'], label="BDE Cache")
pf, = ax.plot(ip['episode_time'], label="IP Cache")
pf, = ax.plot(bde_ip['episode_time'], label="BDE + IP Cache")
pf, = ax.plot(no_cache['episode_time'], label="No Cache (Still Running)")

ax.set_xlabel("Episode")
ax.set_ylabel("Computation Time (Second)")
plt.title("Computation Time (Second): Max ETKDG Attempts = 128")
plt.legend()

plt.savefig("./Results/etkdg_128_episode_time.png", dpi = 500)
plt.show()
```
