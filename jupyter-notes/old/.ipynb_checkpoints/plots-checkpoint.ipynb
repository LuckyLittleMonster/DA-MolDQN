```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd
import pickle
import glob
import json
import hyp
from rdkit import Chem
from rdkit.Chem import QED
from matplotlib.ticker import FuncFormatter


average_episodes = 1
plog = False

def remove_invaild_reward(r):
    if plog:
        return r
#     return r
    if r < 0.0:
        return 0.0
    if r > 3.0:
        return 0.0
    return r
def add_reward(rs, path):
    reward_list = np.load(path)
    final_reward = np.array(reward_list[-average_episodes:])
    final_reward = np.average(final_reward, axis=0)
    for r in final_reward:
        rs.append(remove_invaild_reward(r))

def add_initial_rewards(trial, world_size, step=1, init_rs = []):
    # init_rs = []
    for r in np.arange(0, world_size, step):
        path = f'./Experiments/trial_{trial}_rank_{r}_initial_reward.pickle'
        with open(path, 'rb') as f:
            init_rewards = pickle.load(f)
            for r in init_rewards:
                init_rs.append(r)
    return init_rs

def add_trial_rewards(trial, world_size, step=1, rs = []):
    for r in np.arange(0, world_size, step):
        path = f'./Experiments/trial_{trial}_rank_{r}_rewards.npy'
        add_reward(rs, path)
    return rs

def add_trial_group_rewards(rs, prefix, a, b ,s, w):
    for t in np.arange(a, b, s):
        for r in np.arange(w):
            path = f'./Experiments/trial_{prefix}{t}_rank_{r}_rewards.npy'
            add_reward(rs, path)
    return rs

def to_GB(m):
    return m / 1024 / 1024/ 1024
def add_trial_max_memory(trial, world_size, step=1):
    memory = []
    for r in np.arange(0, world_size, step):
        path = f'./Experiments/trial_{trial}_rank_{r}_memory.npy'
        memory.append(np.load(path))
    memory = np.array(memory)
    total_memory = np.sum(memory, axis = 0)
    return to_GB(np.max(total_memory))
# print(add_trial_max_memory(573, 64))
def add_trial_group_max_memory(ms, prefix, a, b ,s, w):
    for t in np.arange(a, b, s):  
        trial = f'{prefix}{t}'
        ms.append(add_trial_max_memory(trial, w))
    return ms
# print(add_trial_group_max_memory([], 2200, 0, 256, 8, 8))
```

```python
def read_ave_of_top20p_rewards(trial, world_size):
    rs = []
    for rank in range(world_size):
        reward_list = np.load('./Experiments/trial_{}_rank_{}_rewards.npy'.format(trial, rank))
        final_reward = reward_list[-1]
        for r in final_reward:
            rs.append(r)
    rs.sort()
    sz = len(rs)//5
    return np.average(rs[-sz:])
trials = range(570, 580)
world_size = 64
# texts = ['12', '10', '8', '7', '6', '5', '4', '3'] 
texts = ['20', '16', '12', '10', '8', '7', '6', '5', '4', '3'] 

rs = []
for trial in trials:
    r = read_ave_of_top20p_rewards(trial, world_size)
    rs.append(r)
plt.figure(figsize=(8,4))
plt.xlabel('Max Steps')
plt.ylabel('Average Rewards of Top 20% Generated Mols')
# plt.ylim((0, max(max(cts),max(cts_base)) + 0.5))
plt.ylim((1.0, max(rs) + 0.25))
b = plt.bar(range(len(rs)), rs, tick_label=texts, color=['orange'])
plt.bar_label(b, label_type='edge', padding = 5, fmt='%.1f')
# plt.xticks(rotation=45)
# plt.legend()
plt.title('Compare Rewards for different max steps (UE = 1, Batch_Size = 512, 256 mols)')
plt.savefig("./Results/Top_20p_Rewards_57x.png", dpi=500)
plt.show()
```

```python
import random
import math

random.seed(0)
map_range_x0, map_range_x1 = -10, 10
map_range_y0, map_range_y1 = -10, 10
radius = min(map_range_x1 - map_range_x0, map_range_y1 - map_range_y0) / 2
num_init_points = 1
trys = 1
steps = 20
step_length = 0.75

def random_circle(r, d):
    return r * math.sin(d), r * math.cos(d)

def random_walk(num_init_points, trys, steps, step_length):
    rx = []
    ry = []
    for _ in range(num_init_points):
        x, y = random_circle(
            # random.uniform(0.0, math.sqrt(radius * radius * 0.8))
            # random.uniform(0.0, (radius ** (1.5) * 0.8) ** (1.0 / 1.5)),
            math.sqrt(random.uniform(0.0, radius * radius * 0.8)),
            random.uniform(0.0, math.radians(360)))
        for _ in range(trys):
            px = [x]
            py = [y]
            # sd = random.uniform(0.0, math.radians(360.0))
            for _ in range(steps):
                r = random.uniform(0.0, step_length);
                sd = math.atan2(px[-1], py[-1]) + 1.57
                d = sd + random.uniform(math.radians(-90.0), math.radians(90.0))
                # d = sd
                nx = px[-1] + r * math.sin(d)
                ny = py[-1] + r * math.cos(d)
                if nx *nx + ny * ny >= radius * radius * 0.95:
                    # out of range
                    break
                px.append(nx)
                py.append(ny)
            rx.append(px)
            ry.append(py)
    return rx, ry

rx, ry = random_walk(num_init_points, trys, steps, step_length)

plt.figure(figsize=(6,6))
plt.xlim((map_range_x0, map_range_x1))
plt.ylim((map_range_y0, map_range_y1))

x = np.linspace( map_range_x0 , map_range_x1 , 150 )
y = np.linspace( map_range_y0 , map_range_y1 , 150 )
a, b = np.meshgrid( x , y )
C = a ** 2 + b ** 2 - 99
smu = plt.contour( a , b , C , [0])

for x, y in zip(rx, ry):
    init_ps = plt.scatter(x[0],y[0],color='blue', s = 4, label = 'Initial Molecules')
    init_ps.set_zorder(3)
    # plt.scatter(x[1:],y[1:],color='red', s = 2).set_zorder(2)
    # plt.plot(x, y, color='pink')[0].set_zorder(1)
    path = plt.plot(x,y, marker='.', markerfacecolor = 'red', markeredgecolor = 'red', color = 'pink', 
             markersize=2, linewidth=1, label = 'Explored Molecules')[0]
    path.set_zorder(2)
    
plt.legend(handles = [init_ps, path], loc='lower left', bbox_to_anchor=(.75, -0.05))
# plt.legend(handles=[smu, init_ps, path])
plt.axis('off')
plt.savefig(f'./Results/random_walk_{num_init_points}.png', dpi=500, bbox_inches = 'tight')
    
```

```python
# relative episode time, cache hit rate, 600
trial = 600
ranks = range(1)
step = 1
# trial = 2200
# ranks = range(8)
# step = 8

def movingaverage(interval, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'valid')

# memory = None
# for t in range(0, 256):
#     temt = f'{trial}{t}'
#     memory_t = [np.load('./Experiments/trial_{}_rank_{}_memory.npy'.format(temt, rank)) for rank in ranks]
#     for mi in memory_t[1:]:
#         memory_t[0] += mi
#     memory_t = memory_t[0] / (1024 * 1024 * 1024)

#     if memory is None:
#         memory = memory_t
#     else:
#         memory = memory + memory_t
# memory = memory/256
# print(len(memory))

# 2200, 0, 256, 8, 8

def get_data(trial, ranks, step):
    episode_time = None
    for t in np.arange(0, 256, step):
        temt = f'{trial}{t}'
        episode_time_t = np.load('./Experiments/trial_{}_rank_0_episode_time.npy'.format(temt))
        if episode_time is None:
            episode_time = episode_time_t
        else:
            episode_time = episode_time + episode_time_t
    episode_time = episode_time / len(np.arange(0, 256, step))
    
    res = episode_time[0] / episode_time
    print(np.max(res))
    
    episode_lru_cache_hit_rate = None
    for t in np.arange(0, 256, step):
        temt = f'{trial}{t}'
        episode_lru_cache_hit_rate_t = [np.load('./Experiments/trial_{}_rank_{}_episode_lru_cache_hit_rate.npy'.format(temt, rank)) for rank in ranks]
        for mi in episode_lru_cache_hit_rate_t[1:]:
            episode_lru_cache_hit_rate_t[0] += mi
        episode_lru_cache_hit_rate_t = episode_lru_cache_hit_rate_t[0] / (len(ranks))
        if episode_lru_cache_hit_rate is None:
            episode_lru_cache_hit_rate = episode_lru_cache_hit_rate_t
        else:
            episode_lru_cache_hit_rate = episode_lru_cache_hit_rate + episode_lru_cache_hit_rate_t
    
    episode_lru_cache_hit_rate = episode_lru_cache_hit_rate / len(np.arange(0, 256, step))
    
    episode_time = movingaverage(episode_time, 20)
    hit_rate = movingaverage(episode_lru_cache_hit_rate, 20)
    
    episodes = episode_time.shape

    return episode_time, hit_rate, episode_time.shape

episode_time_600, hit_rate_600, episodes = get_data(600, range(1), 1)
episode_time_2200, hit_rate_2200, episodes = get_data(2200, range(8), 8)

fig, (ax1, ax2) = plt.subplots(2,1, figsize=(6,4), sharey=True)

ax12 = ax1.twinx()
p1e, = ax1.plot(episode_time_600, "b-", label = "Episode Time")
p1h, = ax12.plot(hit_rate_600, "r-", label=f"BDE Cache Hit Rate")
ax1.set_ylim(0, 3)
ax12.set_ylim(0, 1.05)
ax1.set_xticklabels([])
ax12.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
ax1.tick_params(axis='y', colors=p1e.get_color())
ax12.tick_params(axis='y', colors=p1h.get_color())





ax22 = ax2.twinx()
p2e, = ax2.plot(episode_time_2200, "b-", label = "Episode Time")
p2h, = ax22.plot(hit_rate_2200, "r-", label=f"BDE Cache Hit Rate")
ax2.set_ylim(0, 7)
ax22.set_ylim(0, 1.05)
ax22.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
ax2.tick_params(axis='y', colors=p2e.get_color())
ax22.tick_params(axis='y', colors=p2h.get_color())



fig.text(0.5, 0.00, 'Episode', va = 'center', ha='center')
fig.text(0.06, 0.5, 'Episode Time (second)', va = 'center', ha='center', rotation='vertical', color=p1e.get_color())
fig.text(1.02, 0.5, 'BDE Cache Hit Rate', va = 'center', ha='center', rotation='vertical', color=p1h.get_color())



# ax1.set_xlabel("Episode")
# ax1.plot(time, data1, color=c1)
# ax1.set_xlabel('time (s)')
# ax1.set_ylabel('exp')
# ax2.plot(time, data2, color=c2)
# ax2.set_ylabel('sin')

# fig, host = plt.subplots(figsize = (6, 2))
# fig.subplots_adjust(right=0.75)

# par1 = host.twinx()
# # par2 = host.twinx()

# # move the spine of the second axes outwards
# # par2.spines["right"].set_position(("axes", 1.2))

# p1, = host.plot(episode_time, "b-", label = "Episode Time")
# p2, = par1.plot(hit_rate, "r-", label=f"BDE Cache Hit Rate")
# # p3, = par2.plot([0, 1, 2], [50, 30, 15], 'b-', label="Velocity")

# # host.set_xlim(0, 2)
# if trial == 600:
#     host.set_ylim(0, 3)
# elif trial == 2200:
#     host.set_ylim(0, 4.5)
    
# par1.set_ylim(0, 1.05)
# # par2.set_ylim(1, 65)

# if trial == 2200:
#     host.set_xlabel("Episode")
# host.set_ylabel("Episode Time (second)")
# par1.set_ylabel('BDE Cache Hit Rate')
# # par2.set_ylabel("Velocity")

# lines = [p1, p2]
# # host.legend(lines, [l.get_label() for l in lines], loc = 'center right')

# for ax in [par1]:
#     ax.set_frame_on(True)
#     ax.patch.set_visible(False)

#     plt.setp(ax.spines.values(), visible=False)
#     ax.spines["right"].set_visible(True)
    
# host.yaxis.label.set_color(p1.get_color())
# par1.yaxis.label.set_color(p2.get_color())
# # par2.yaxis.label.set_color(p3.get_color())

# # host.spines["right"].set_edgecolor(p1.get_color())
# # par1.spines["right"].set_edgecolor(p2.get_color())
# # par2.spines["right"].set_edgecolor(p3.get_color())
# par1.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 


# host.tick_params(axis='y', colors=p1.get_color())
# par1.tick_params(axis='y', colors=p2.get_color())
# # par2.tick_params(axis='y', colors=p3.get_color())
# # 
# if trial==600:
#     plt.xticks([], [])
plt.suptitle('Episode Time & BDE Cache Hit Rate')
plt.legend([p1e, p1h],
           ['Episode Time','BDE Cache Hit Rate'], 
           loc='lower center', bbox_to_anchor=(0.5,-0.7),
           ncol=2
          )

plt.savefig("./Results/Episode_Time_Cache_600_2200.png".format(trial), dpi = 500, bbox_inches = 'tight')
plt.show()
```

```python
def two_scales(ax1, time, data1, data2, c1, c2):
    ax2 = ax1.twinx()
    ax1.plot(time, data1, color=c1)
    ax1.set_xlabel('time (s)')
    ax1.set_ylabel('exp')
    ax2.plot(time, data2, color=c2)
    ax2.set_ylabel('sin')
    return ax1, ax2

# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
s2 = np.sin(2 * np.pi * t)

# Create axes
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))
ax1, ax1a = two_scales(ax1, t, s1, s2, 'r', 'b')
ax2, ax2a = two_scales(ax2, t, s1, s2, 'gold', 'limegreen')

# Change color of each axis
def color_y_axis(ax, color):
    """Color your axes."""
    for t in ax.get_yticklabels():
        t.set_color(color)

color_y_axis(ax1, 'r')
color_y_axis(ax1a, 'b')
color_y_axis(ax2, 'gold')
color_y_axis(ax2a, 'limegreen')

plt.tight_layout()
plt.show()
```
