```python
def mol_with_atom_index(mol):
    for atom in mol.GetAtoms():
        atom.SetAtomMapNum(atom.GetIdx())
    return mol

def movingaverage(interval, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'valid')
```

```python
def make_prediction(model, molbert_fn, smiles):
    """
    Predictions of a property for a smile. 
    Details: Convert smile to molbert features, apply the model to the molbert features. 
    It returns a the list of predictions
    :param model: the input model for the prediction of the property of interrest
    :param molbert_fn: the function to transform smile to features
    :param smiles a smile or a list of smiles we want to apply the model on
    :return array with the predictions. 
    """
    if type(smiles)!=list:
        smiles=[smiles]
    s_features = np.vstack([molbert_fn(batch) for batch in batchify(smiles, 256)])
    return model.predict(s_features)
```

```python
def molbert_transform(smile):
    original_checkpoint='./TotalBenchmark/model_and_data_for_dai/molbert_100epochs/checkpoints/last.ckpt'
    molbert = MolBertFeaturizer(original_checkpoint, 
                            embedding_type='average-1-cat-pooled', 
                            max_seq_len=200)
    return molbert.transform([smile])
```

```python
original_checkpoint='./TotalBenchmark/model_and_data_for_dai/molbert_100epochs/checkpoints/last.ckpt'
molbert = MolBertFeaturizer(original_checkpoint, 
                            embedding_type='average-1-cat-pooled', 
                            max_seq_len=200)  # type: ignore
molbert_fn = lambda smiles: molbert.transform(smiles)[0]

# load the bde model
bde_model = joblib.load('./TotalBenchmark/model_and_data_for_dai/models/bde_molbertmodel.pkl') # reminder: goal small bde

# load the ip model
ip_model = joblib.load('./TotalBenchmark/model_and_data_for_dai/models/ip_molbertmodel.pkl') # reminder: goal high ip

scaler_bde= joblib.load('./TotalBenchmark/model_and_data_for_dai/models/scaler_bde.pkl')
scaler_ip= joblib.load('./TotalBenchmark/model_and_data_for_dai/models/scaler_ip.pkl')

def batchify(iterable, batch_size):
    for ndx in range(0, len(iterable), batch_size):
        batch = iterable[ndx: min(ndx + batch_size, len(iterable))]
        yield batch
        
def make_prediction(model, molbert_fn, smiles):
    """
    Predictions of a property for a smile. 
    Details: Convert smile to molbert features, apply the model to the molbert features. 
    It returns a the list of predictions
    :param model: the input model for the prediction of the property of interrest
    :param molbert_fn: the function to transform smile to features
    :param smiles a smile or a list of smiles we want to apply the model on
    :return array with the predictions. 
    """
    if type(smiles)!=list:
        smiles=[smiles]
    s_features = np.vstack([molbert_fn(batch) for batch in batchify(smiles, 256)])
    return model.predict(s_features)


from sklearn import preprocessing

def score_from_one_prediction(bde, ip, scaler_bde, scaler_ip):
    """
    Proposed score (to validate with Sylvain and Sophie.)
    :param bde float: bde value for the tested smile
    :param ip float: ip value for the tested smile
    :param scaler_bde: scaler trained on bde dataset
    :param scaler_ip: scaler trained on the ip dataset
    :return score
    """
    bde = scaler_bde.transform([[bde]])
    ip = scaler_ip.transform([[ip]])
    # score between 0 and 1 ??? => to discuss with DAI
    return ip + (1-bde)
```

```python
weight = 0.2
key = list(valid_actions.keys())[0]

QED_list = (valid_actions[key])[0::2]
mol_list = (valid_actions[key])[1::2]
SA_score_list = [sascorer.calculateScore(mol) for mol in mol_list]
SA_QED_list = [qed - weight*sascorer.calculateScore(mol) for mol,qed in zip(mol_list,QED_list)]
mol_list = [Chem.MolToSmiles(mol) for mol in mol_list]

mol_list_comb = zip(SA_QED_list, QED_list, mol_list, SA_score_list)
# Remove the duplicated elements
mol_list_comb = list(set(mol_list_comb))
mol_list_sort = sorted(mol_list_comb, key = lambda x: x[0])
```

```python
SA_score_list = [score for (_,_,_,score) in mol_list_sort]
QED_list = [qed for (_,qed,_,score) in mol_list_sort]
SA_QED_list = [qed-weight*score for (_,qed,_,score) in mol_list_sort]
smile_list = [mol for (_,_,mol,score) in mol_list_sort]


ex_smile=key
qed = QED.qed(Chem.MolFromSmiles(ex_smile))
SA_score = sascorer.calculateScore(Chem.MolFromSmiles(ex_smile))
SA_QED = qed-0.2*SA_score
```

```python
table = BeautifulTable()
table.rows.append(QED_list[-6:])
table.rows.append(SA_score_list[-6:])
table.rows.append(SA_QED_list[-6:])
table.columns.append([qed, SA_score, SA_QED])
table.rows.header = ["QED", "SA_score", "reward"]
table.columns.header = ["mol 1", "mol 2", "mol 3", "mol 4", "mol 5", "mol 6", "init mol"]
# print(table)



QED_list = [round(x, 3) for x in QED_list]
SA_score_list = [round(x, 3) for x in SA_score_list]
SA_QED_list = [round(x, 3) for x in SA_QED_list]


table = PrettyTable()
# table.title = "Train on a multi-task model, Epsilon Decay=0.999, Init Epsilon=0.1"
table.title = "Train on a single molecule MolDQN, Epsilon Decay=0.999, Init Epsilon=0.1"
table.field_names = ["", "mol 1", "mol 2", "mol 3", "mol 4", "mol 5", "mol 6", "init mol"]
table.add_row(["QED"] + QED_list[-6:] + [round(qed, 3)])
table.add_row(["SA_score"] + SA_score_list[-6:] + [round(SA_score, 3)])
table.add_row(["Reward"] + SA_QED_list[-6:] + [round(SA_QED, 3)])
print(table)
    
Draw.MolsToGridImage([Chem.MolFromSmiles(mol[2]) for mol in mol_list_sort[-6:]], molsPerRow=3, subImgSize=(250, 120))
```
